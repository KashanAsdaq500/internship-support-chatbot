```python
# ==============================
# Internship Support Chatbot
# NLP FAQ Assistant
# ==============================

# Install required library
!pip -q install sentence-transformers

from sentence_transformers import SentenceTransformer, util

# Load NLP model
model = SentenceTransformer('all-MiniLM-L6-v2')

# -----------------------------
# STEP 1: FAQ DATA
# -----------------------------

faq = {
    "What is the internship duration?":
        "The internship duration is 3 months.",

    "How do I submit my tasks?":
        "Tasks should be submitted through the internship portal.",

    "Who do I contact for technical issues?":
        "You can contact the IT support team or your supervisor.",

    "What is the attendance requirement?":
        "Interns must maintain at least 80% attendance.",

    "Will I receive a certificate?":
        "Yes, certificates are awarded after successful completion."
}

questions = list(faq.keys())
answers = list(faq.values())

# Convert questions into vectors
question_embeddings = model.encode(questions, convert_to_tensor=True)

# -----------------------------
# STEP 2: CHAT FUNCTION
# -----------------------------

def chatbot(user_query):
    query_embedding = model.encode(user_query, convert_to_tensor=True)

    similarity = util.cos_sim(query_embedding, question_embeddings)
    best_match = similarity.argmax()

    return answers[best_match]

# -----------------------------
# STEP 3: CHAT LOOP
# -----------------------------

print("Internship Support Chatbot")
print("Type 'exit' to stop\n")

while True:
    query = input("You: ")

    if query.lower() == "exit":
        print("Chatbot: Goodbye! 👋")
        break

    response = chatbot(query)
    print("Chatbot:", response)
```
